## 1. Introduction
Numerical computation involves solving mathematical problems by representing numbers as approximations (floating-point arithmetic) rather than exact symbolic expressions. In this course, we use **Python** and specifically the library **NumPy** (*Numerical Python*) because it is highly optimized for vector and matrix operations.

### Why NumPy over Lists?
* **Performance:** NumPy arrays are stored in contiguous memory blocks.
* **Functionality:** Built-in linear algebra, Fourier transforms, and random number capabilities.
* **Syntax:** Mathematical operations apply element-wise automatically.

```python
import numpy as np

# A standard Python list
py_list = [1, 2, 3, 4, 5]

# Creating a NumPy array
np_arr = np.array(py_list)

print(f"Python List: {py_list}")
print(f"NumPy Array: {np_arr}")
print(f"Type of np_arr: {type(np_arr)}")
```

## 2. Creating Arrays
We rarely type out arrays manually in numerical physics. Instead, we generate them using functions like `arange`, `linspace`, or `zeros`.

* `np.arange(start, stop, step)`: Like Python's `range` but returns an array.
* `np.linspace(start, stop, num)`: Generates `num` points evenly spaced between `start` and `stop` (Crucial for plotting and grids).
* `np.zeros(N)` / `np.ones(N)`: Creates arrays filled with 0s or 1s.

```python
# Create an array of integers from 0 to 9
arr_range = np.arange(0, 10, 1.3)
print(f"Arange: {arr_range}")

print(type(arr_range))
print(arr_range.shape)

print(arr_range.reshape(4,-1).T)

```

## 3. Data Types
Unlike Python lists, a NumPy array must contain data of the **same type**. This allows the computer to optimize memory usage. Common types in physics:

* `int32` / `int64`: Integers
* `float32` / `float64`: Floating point numbers (decimals)
* `complex128`: Complex numbers ($a + bi$)

```python
# Integer array
int_arr = np.array([1, 2, 3], dtype=np.int64)
print(f"Integer Array dtype: {int_arr.dtype}")

# Float array (Standard for scientific computing)
float_arr = np.array([1., 2., 3.], dtype=np.float64)
print(f"Float Array dtype: {float_arr.dtype}")

# Complex array
complex_arr = np.array([1+2j, 3+4j])
print(f"Complex Array:\n{complex_arr}")
print(f"Real part: {complex_arr.real}")
print(f"Imaginary part: {complex_arr.imag}")
```

## 4. Precision and Floating Point Arithmetic
Computers cannot store infinite precision. They store numbers in scientific notation (binary).
* **Machine Epsilon:** The smallest difference between $1.0$ and the next representable floating-point number.
* **Round-off Error:** The error introduced when a number cannot be exactly represented.

**Key Example:** $0.1 + 0.2$ often does not equal exactly $0.3$ in floating point arithmetic.

```python
# The classic floating point issue
val = 0.1 + 0.2
print(val)
print(f"0.1 + 0.2 = {val:.4f}")
print(f"Is it exactly 0.3? {val == 0.3}")

# Checking Machine Epsilon for standard float (64-bit)
# This is the 'resolution' of our number system
eps = np.finfo(float).eps
print(f"\nMachine Epsilon (float64): {eps}")

# Comparing 32-bit vs 64-bit precision
# Notice how 32-bit loses accuracy faster
float32_eps = np.finfo(np.float32).eps
print(f"Machine Epsilon (float32): {float32_eps}")
```
