# Introduction to Sorting



## Introduction to Sorting Algorithms

Sorting is a fundamental operation in computer science that arranges elements of a list or an array in a specific order (e.g., numerical, alphabetical). Efficient sorting is crucial for optimizing other algorithms like searching and merging, improving data readability, and enabling advanced data structures.

### 1. Bubble Sort

**Basic Logic:** Bubble Sort works by repeatedly stepping through the list, comparing adjacent elements, and swapping them if they are in the wrong order. This process is repeated until no swaps are needed, indicating that the list is sorted. Larger elements 'bubble' to the end of the list with each pass.

**Time Complexity:** $O(N^2)$ in the worst and average cases, where N is the number of elements. This makes it inefficient for large datasets. It might be used for very small datasets or as a pedagogical tool due to its simple implementation.

### 2. Quick Sort

**Basic Logic:** Quick Sort is a divide-and-conquer algorithm. It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. The sub-arrays are then recursively sorted. This process repeats until the entire array is sorted.

**Time Complexity:** On average, Quick Sort has a time complexity of $O(N \log N)$, which is significantly more efficient than $O(N^2)$ for larger datasets. In the worst-case scenario, its complexity can degrade to $O(N^2)$, though this is rare with good pivot selection strategies.

### Python's `numpy.sort`

Python's `numpy.sort` function is highly optimized for performance. It typically uses advanced sorting algorithms like **QuickSort** or **MergeSort** (or a hybrid approach like Timsort) internally, depending on the data type and other factors. This makes `numpy.sort` very fast and efficient for sorting large numerical arrays.

## Manual Bubble Sort Implementation


**Reasoning**:
This code block defines the function, implement the sorting logic, and then show an example of how to use it with an unsorted list.



````python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


# Example usage:

my_list = [64, 34, 25, 12, 22, 11, 90, 44, 58]

print("Original list:", my_list)


bubble_sort(my_list)

print("Sorted list:", my_list)
````

## NumPy Sorting Implementation

This Code cell demonstrates how to use `np.sort()` for sorting values and `np.argsort()` for obtaining sorted indices. It includes an explanation of why `argsort` is particularly useful in scientific contexts (e.g., sorting one array based on another).


````python
import numpy as np

# 2. Create a sample NumPy array
data = np.array([5, 2, 8, 1, 9, 3])

print("Original array:", data)

# 3. Demonstrate np.sort()
sorted_data = np.sort(data)[::-1]
print("Array sorted with np.sort():", sorted_data)
print("Original array after np.sort() (unchanged):", data)

# 4. Demonstrate np.argsort()
sorted_indices = np.argsort(data, )
print("Indices that would sort the array (np.argsort()):", sorted_indices)

# Use sorted_indices to reorder the original data array
sorted_using_indices = data[sorted_indices]
print("Array sorted using np.argsort() indices:", sorted_using_indices)
````

### Explanation of `np.argsort()` Utility

`np.argsort()` is particularly powerful because it returns the *indices* that would sort an array, rather than the sorted array itself. This allows for flexible reordering of data. Its usefulness shines in scientific and data analysis contexts, especially when dealing with multiple related arrays.

For example, imagine you have two related arrays:
*   `names = np.array(['Alice', 'Bob', 'Charlie', 'David'])`
*   `scores = np.array([85, 92, 78, 90])`

If you want to sort the `names` based on `scores`, you would use `argsort` on the `scores` array to get the sorting order, and then apply that order to both `names` and `scores` to keep their correspondence intact:

```python
import numpy as np

names = np.array(['Alice', 'Bob', 'Charlie', 'David'])
scores = np.array([85, 92, 78, 90])

sort_order = np.argsort(scores) # Get indices that sort scores

sorted_names = names[sort_order]
sorted_scores = scores[sort_order]

print("Sorted Names:", sorted_names)
print("Sorted Scores:", sorted_scores)
# Expected Output:
# Sorted Names: ['Charlie' 'Alice' 'David' 'Bob']
# Sorted Scores: [78 85 90 92]
```

This maintains the relationship between `names` and `scores` while sorting by one of them, which is a common requirement in data manipulation and analysis.
